```python
import xarray as xr
import numpy as np
from remap.cubesphere import to_chunked_dataset

ds = xr.open_dataset('/home/disk/rhodium/bowenliu/ERA5/era5_1950-2022_3h_1deg_z10.nc')
ds
```

```python
f1 = ds.squeeze().z[0,...].squeeze()
f2 = ds.squeeze().z[1,...].squeeze()
f3 = ds.squeeze().z[2,...].squeeze()
f4 = ds.squeeze().z[3,...].squeeze()

ds_all = xr.concat([f1, f2, f3, f4], dim='valid_time')
ds_all
```

```python
ds_all_clean = ds_all.dropna(dim='valid_time',how='any').sortby('valid_time')
ds_all_clean = ds_all_clean.rename({'valid_time':'time'})
```

```python
ds_all_clean = xr.Dataset({'z':ds_all_clean})
ds_all_clean
```

```python
ds_all_clean.to_netcdf('/home/disk/rhodium/bowenliu/ERA5/era5_1950-2022_3h_1deg_z150.nc')
```

```python
ds = ds.rename({'valid_time':'time'}).squeeze()
ds
```

```python
ds.to_netcdf('/home/disk/rhodium/bowenliu/ERA5/era5_1950-2022_3h_1deg_z50.nc')
```

```python
nside = 32
const = False if "time" in list(list(ds.dims.keys())) else True

coords = {}
if not const:
    coords["time"] = ds.coords["time"]
coords["face"] = np.array(range(12), dtype=np.int64)
coords["height"] = np.array(range(nside), dtype=np.int64)
coords["width"] = np.array(range(nside), dtype=np.int64)
chunksizes = {coord: len(coords[coord]) for coord in coords}
chunksizes["time"] = 1
```

```python
ds_hpx = to_chunked_dataset(ds=ds, chunking=chunksizes)
```

```python
ds_hpx.to_netcdf('/home/disk/rhodium/bowenliu/HPX32/era5_1deg_3h_HPX32_1950-2022_z150.nc')
```
